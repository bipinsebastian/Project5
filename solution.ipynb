```python
import matplotlib.pyplot as plt
import pandas as pd

# Count the number of songs per artist
music_data = pd.read_csv(r'C:\Users\Gymnasiet\floeder for python\data-hw-bipinsebastian\projects\project 5\spotify_songs.csv')
#print(music_data.head())
top_artists = music_data['track_artist'].value_counts().nlargest(10)

# Plot
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='skyblue')
plt.title('Top 10 Artists with the Most Songs')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

2. Show in a graph the top 10 most popular songs

```python
top_popular_songs = music_data[['track_album_name','track_popularity']].sort_values(by='track_popularity', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(top_popular_songs['track_album_name'], top_popular_songs['track_popularity'], color='green')
plt.title('Top 10 Most Popular Songs')
plt.xlabel('Song')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. Show in a graph the 10 least popular songs

```python
least_popular_songs = music_data[['track_album_name','track_popularity']].sort_values(by='track_popularity').head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(least_popular_songs['track_album_name'], least_popular_songs['track_popularity'], color='red')
plt.title('Top 10 Least Popular Songs')
plt.xlabel('Song')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

4. Show in a graph the number of albums that were released per year between 2000 and 2023

```python
# Convert 'track_album_release_date' to datetime
music_data['track_album_release_date'] = pd.to_datetime(music_data['track_album_release_date'], errors='coerce')

# Extract the year from 'track_album_release_date'
music_data['release_year'] = music_data['track_album_release_date'].dt.year

# Filter data for albums released between 2000 and 2023
albums_per_year = music_data[music_data['release_year'].between(2000, 2023)]['release_year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
albums_per_year.plot(kind='bar')
plt.title('Number of Albums Released Per Year (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Albums')
plt.show()

```

5. Show in a graph the number of albums that the top 10 artists released

```python
# Assume that 'album' is unique for each artist and use that to count albums
top_artists_albums = music_data.groupby('track_artist')['track_album_name'].nunique().nlargest(10)

# Plot
plt.figure(figsize=(10, 6))
top_artists_albums.plot(kind='bar', color='orange')
plt.title('Number of Albums Released by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Number of Albums')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

6. Show in a graph the longest 10 songs

```python
longest_songs = music_data[['track_album_name', 'duration_ms']].sort_values(by='duration_ms', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(longest_songs['track_album_name'], longest_songs['duration_ms'], color='brown')
plt.title('Top 10 Longest Songs')
plt.xlabel('Song')
plt.ylabel('Duration (seconds)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

7. Show in a graph the 10 most danceable songs of artists with the most songs

```python
# Get top 10 artists with the most songs
top_artists = music_data['track_artist'].value_counts().nlargest(10).index

# Filter songs by these artists and get the most danceable ones
most_danceable_songs = music_data[music_data['track_artist'].isin(top_artists)].sort_values(by='danceability', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(most_danceable_songs['track_album_name'], most_danceable_songs['danceability'], color='teal')
plt.title('Top 10 Most Danceable Songs of Artists with Most Songs')
plt.xlabel('Song')
plt.ylabel('Danceability')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

8. Show in a graph the 10 loudest and most popular songs

```python
loudest_popular_songs = music_data.sort_values(by=['loudness', 'track_popularity'], ascending=[False, False]).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(loudest_popular_songs['track_album_name'], loudest_popular_songs['loudness'], color='blue')
plt.title('Top 10 Loudest and Most Popular Songs')
plt.xlabel('Song')
plt.ylabel('Loudness (dB)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
